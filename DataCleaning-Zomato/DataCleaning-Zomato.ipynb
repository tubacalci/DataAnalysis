```python
colunmns_to_keep = ['name', 'online_order', 'book_table', 'rate', 'dish_liked','approx_cost(for two people)']
columns_to_drop = ['url', 'address', 'votes','phone', 'location', 'rest_type','cuisines','reviews_list', 'menu_item',
       'listed_in(type)', 'listed_in(city)']
```

# Deleting Redundant Columns

## 2nd Method - Dropping the Columns

```python
columns_to_drop
```

# Renaming the Columns

```python
data
```

```python
data.columns[0].capitalize()
```

# Dropping Duplicates

```python
data.duplicated().sum()
```

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().sum()
```

# Remove the NaN Values from the Dataset

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
data
```

```python
data.reset_index(inplace = True)
data
```

```python
data.drop(columns = "index", inplace = True)
data
```

# Cleaning Individual Columns

```python
data["Rate"]
```

```python
data["Rate"] = data["Rate"].str.replace(" ", "")
data["Rate"]
```

```python
data["Rate"] = data["Rate"].str.replace("/", " out of ")
```

```python
data["Rate"]
```

```python
data
```

# Check for some more Transformations

```python
data["Online_order"]
```

```python
data["Online_order"] = data["Online_order"].map({"Yes": 1, "No": 0})
```

```python
data
```

```python
data["Book_table"] = data["Book_table"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
data
```

# Export Final Cleaned Dataset

```python
data.to_csv("cleaned_zomato_data.csv", index = False)
```
